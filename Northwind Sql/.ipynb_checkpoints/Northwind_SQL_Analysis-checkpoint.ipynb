## Ranking Employee Sales Performance

We will comprehensively review the company's sales performance from an employee perspective. The objective is twofold:

* First, the management team wants to recognize and reward top-performing employees, fostering a culture of excellence within the organization.
* Second, they want to identify employees who might be struggling so that they can offer the necessary training or resources to help them improve.

The management team is keen on encouraging healthy competition and rewarding stellar performers. Therefore we would rank employees based on their total sales amount.

```python
query = """
WITH emp_tot_sales AS (
SELECT e.employee_id, e.first_name|| ' ' ||e.last_name AS emp_name,e.title, SUM(od.unit_price*od.quantity*(1-od.discount)) AS tot_sales
  FROM employees AS e
  JOIN orders AS o
    ON e.employee_id=o.employee_id
  JOIN order_details AS od
    ON o.order_id=od.order_id                    
 GROUP BY e.employee_id,emp_name,e.title
)

SELECT employee_id, emp_name, title, tot_sales, RANK() OVER(ORDER BY tot_sales DESC) AS emp_rank
  FROM emp_tot_sales
"""

result = %sql $query

emp_sales = pd.DataFrame(result)
emp_sales
```

From the table, it can be seen that `Margaret Peacock`, `Janet Leverling` and `Nancy Davolio` are the top 3 performing employees at Northwind Traders. These employees should be incentivised and rewarded with appropriate bonus so as to foster a culture of excellence within the organization. 

On the contrary, various training and/or counseling programs should be arranged for `Anne Dodsworth`, `Micheal Suyama` and `Steven Buchanan`to address the inefficiencies in their performance, which would help them to perform better. 

## Running Total of Monthly Sales

After gaining insight into Employee performance, The Management is now keen on gaining a more macro-level perspective, specifically around the company's overall sales performance over time. They are now looking to visualize the progress of the sales and identify trends that might shape the company's future strategies.

Therefore in this new analysis, we will visualize the company's sales progress over time on a monthly basis. This will involve aggregating the sales data at a monthly level and calculating a running total of sales by month. This visual will provide the management team with a clear depiction of sales trends and help identify periods of high or low sales activity.

```python
query="""
WITH sales_by_month AS (
    SELECT DATE_TRUNC('month',o.order_date) AS order_month, SUM(od.unit_price*od.quantity*(1-od.discount)) AS monthly_sales
      FROM orders AS o
      JOIN order_details AS od
        ON o.order_id=od.order_id
    GROUP BY order_month
    )

SELECT order_month,ROUND(CAST(monthly_sales AS numeric),2) AS monthly_sales, ROUND(CAST(SUM(monthly_sales) OVER(ORDER BY order_month) AS numeric),2) AS running_total
  FROM sales_by_month;
"""

result= %sql $query
mnth_run_sales=pd.DataFrame(result)

fig,ax1=plt.subplots(figsize=(15,6))
ax1.bar(mnth_run_sales['order_month'],mnth_run_sales['monthly_sales'],width=20, label="Monthly Sales",zorder=5)
ax1.set(ylim=(0,mnth_run_sales['monthly_sales'].max()+50000),ylabel="Monthly Sales",xlabel="Date")
ax1.yaxis.set_major_formatter('${x:,.0f}')
ax2=ax1.twinx()
ax2.plot(mnth_run_sales['order_month'],mnth_run_sales['running_total'],color="orange",label="Running Total",marker='o')
ax2.set_yticks([500000,1000000],[500000,1000000])
ax2.set(ylabel="Running Total")
ax2.yaxis.set_major_formatter('${x:,.0f}')
handles1, labels1=ax1.get_legend_handles_labels()
handles2, labels2=ax2.get_legend_handles_labels()
plt.legend(handles=handles1 + handles2, labels=labels1 + labels2, loc='upper left')
ax1.grid(zorder=0)

plt.show()
```

The above plot depicts the YoY increasing trend in monthly sales figures. The increasing trend can be attributed to the growing popularity of Northwind Traders. 

It can be inferred that Northwind Traders is currently in its "Growth" phase. Therefore as per the strategic aspect, Management should look forward to capitalize on its "Growth" phase by expanding its business by increasing its brick-and-mortar stores along with an increase in its online presence. 

## Month-Over-Month Sales Growth

After presenting the running sales total by month, the management team is interested in further dissecting these figures. They would like to analyze the month-over-month sales growth rate. Understanding the rate at which sales are increasing or decreasing from month to month will help the management team identify significant trends.

```python
query="""
WITH monthly_sales AS (
    SELECT DATE_TRUNC('month',o.order_date) AS order_month, SUM(od.unit_price*od.quantity*(1-od.discount)) AS month_sales
      FROM orders AS o
      JOIN order_details AS od
        ON o.order_id=od.order_id
    GROUP BY order_month
    ),

prev_month_sales AS (
    SELECT order_month, month_sales, LAG(month_sales) OVER(ORDER BY order_month) AS prev_month_sales
    FROM monthly_sales
)

SELECT order_month,month_sales, ((month_sales-prev_month_sales)/prev_month_sales)*100 AS growth_rate
  FROM prev_month_sales
"""

result= %sql $query
mom_sales=pd.DataFrame(result)

fig,ax1 = plt.subplots(figsize=(15,6))
ax1.bar(mom_sales['order_month'],mom_sales['month_sales'],width=20,zorder=5,label="Monthly Sales")
ax1.set(ylim=(0,mom_sales['month_sales'].max()+10000),ylabel="Monthly Sales",xlabel="Date")
ax1.yaxis.set_major_formatter('${x:,.0f}')
plt.grid(zorder=0)

ax2=ax1.twinx()
ax2.plot(mom_sales['order_month'],mom_sales['growth_rate'],color="orange",marker='o',label="Growth Rate %")
ax2.set(ylim=(-100,100),ylabel="Growth Rate %")
ax2.yaxis.set_major_formatter('{x:,.2f}%')
plt.title("Monthly Sales and Month-On-Month Growth Rate (%)")
handles1, labels1=ax1.get_legend_handles_labels()
handles2, labels2=ax2.get_legend_handles_labels()
plt.legend(handles=handles1 + handles2, labels=labels1 + labels2, loc='upper left')
plt.show()
```

From the above plot, it can be seen that the Monthly sales figures declined during `August-1996`,`February-1997`,`June-1997`, `August-1997` and  `November-1997`. We will disregard the figures for `May-1998` because an outlier decline in growth rate signifies that the data for this month and year combination is not present in its entirety. 

We don't have sufficient data to infer any kind of seasonality from this data. Further data collection needs to be done for a more comprehensive analysis. However, as per available data, a further drill-down needs to be performed to arrive at the reasoning of negative growth rates in the time periods mentioned above.

## Identifying High-Value Customers

Upon presenting the sales growth and trend analysis, The Management want to identify high-value customers to whom they can offer targeted promotions and special offers, which could drive increased sales, improve customer retention, and attract new customers. 

To achieve this, we will classify our customers to be either "Above Average" or "Average/Below Average". Our focus will solely be on customers with "Above Average" order values.

```python
%%sql
WITH cust_orders AS (
    SELECT c.company_name,c.contact_name, SUM(od.unit_price*od.quantity*(1-od.discount)) AS order_val
      FROM orders AS o
      JOIN order_details AS od
        ON o.order_id=od.order_id
      JOIN customers AS c
        ON o.customer_id=c.customer_id
     GROUP BY c.company_name, c.contact_name
),
cust_class AS (
SELECT company_name, contact_name, CASE WHEN order_val>AVG(order_val) OVER() THEN 'Above Average' ELSE 'Average/Below Average' END AS ord_status
  FROM cust_orders
)

SELECT company_name, contact_name
  FROM cust_class
 WHERE ord_status='Above Average'
```

From the above table, it can be seen that there are `30` high-value customers. From a strategic standpoint, The Management at Northwind Traders should allocate a dedicated customer support team to these 30 high-value accounts. Since these 30 accounts are a major driver of Northwind's revenue, it becomes essential to provide them priority customer support, so that they do not face any issue and if they do, their issues are resolved at the earliest.

A further drill down is performed to see the proportion of "Above Average" and "Average/Below Average" orders.

```python
query="""
WITH cust_orders AS (
    SELECT o.customer_id,o.order_id, SUM(od.unit_price*od.quantity*(1-od.discount)) AS order_val
      FROM orders AS o
      JOIN order_details AS od
        ON o.order_id=od.order_id
     GROUP BY o.customer_id,o.order_id
),

ord_status AS (
SELECT order_id, CASE WHEN order_val>AVG(order_val) OVER() THEN 'Above Average' ELSE 'Average/Below Average' END AS ord_status
  FROM cust_orders
)

SELECT ord_status, COUNT(ord_status) AS count_ord_status
  FROM ord_status
 GROUP BY ord_status;
 """

result= %sql $query
ord_class_count=pd.DataFrame(result)
ord_class_count

plt.pie(ord_class_count['count_ord_status'],autopct='%.1f%%')
plt.legend(labels=ord_class_count['ord_status'],loc=8,bbox_to_anchor=(0.5,-0.1))
plt.title("Order Classification")
plt.show()
```

As can be seen from the above pie chart, The number of `Average/Below Average` orders are almost twice those of `Above Average` orders. From a strategic standpoint, Northwind traders should run a marketing campaign, along with various incentives for customers to stimulate the buying behaviour to increase the proportion of `Above Average` orders.

### Top 20 percentage of customers by Purchase Volume

```python
%%sql
WITH cust_purc_vol AS (
    SELECT c.company_name, c.contact_name, SUM(od.quantity) AS tot_purc_vol, PERCENT_RANK() OVER(ORDER BY SUM(od.quantity) DESC) AS rank
      FROM customers AS c
      JOIN orders AS o
        ON c.customer_id=o.customer_id
      JOIN order_details AS od
        ON o.order_id=od.order_id
     GROUP BY c.company_name, c.contact_name
     ORDER BY tot_purc_vol DESC
)

SELECT company_name, contact_name, tot_purc_vol
  FROM cust_purc_vol
 WHERE rank<=0.2;
```

Due to their repeated purchasing behaviour, these customers can be termed as loyal customers. Therefore, along with High-Value accounts, these accounts (There are many accounts common amongst the two) should also be serviced with exclusive and responsive customer support.

## Percentage of Sales for Each Category

Next we provide the management team with an understanding of sales composition across different product categories. By knowing the percentage of total sales for each product category, they can gain insights into which categories drive most of the company's sales.

This understanding will help guide decisions about inventory (e.g., which categories should be stocked more heavily) and marketing strategies (e.g., which categories should be promoted more aggressively).

```python
query="""
WITH cat_tot_sales AS (
  SELECT c.category_name, SUM(od.unit_price*od.quantity*(1-od.discount)) AS tot_sales
    FROM categories AS c
    JOIN products AS p
      ON c.category_id=p.category_id
    JOIN order_details AS od
      ON p.product_id=od.product_id
   GROUP BY c.category_name
)

SELECT category_name, ROUND(CAST((tot_sales/SUM(tot_sales) OVER())*100 AS numeric),2) AS pct_sales
  FROM cat_tot_sales
"""

result= %sql $query
cat_sales_pct=pd.DataFrame(result)

plt.pie(cat_sales_pct['pct_sales'],autopct='%.1f%%')
plt.legend(labels=cat_sales_pct['category_name'],loc=8,bbox_to_anchor=(0.5,-0.4))
plt.title("Percentage of Sales for Each Category")
plt.show()
```

From the above pie chart, it can be seen that the Top 3 categories by sales proportion are `Beverages (21.2%)`, `Dairy Products (18.5%)` and `Confections (13.2%)`, which indicates that products from these categories should be stocked heavily and should never fall below the reorder level, which can lead to shortages and can further result in customer dissatisfaction.

In addition to the top 3 categories mentioned above, inventory considerations regarding products from `Meat/Poultry (12.9%)` and `Seafood (10.4%)` categories should also be made, since these categories also have substantial contributions in total sales. 

On the other hand, aggressive marketing tactics, such as Discounts, Coupons/Vouchers, and advertisements regarding the same should be done for products belonging to `Grains/Cereals (7.65%)`, `Produce (7.9%)`, `Condiments (8.4%)` to increase their proportion amongst total sales. 

## Top Products Per Category

The next step is to drill down further into each group. The management team can keep a track of top three items sold in each product category. This will allow them to identify star performers and ensure that these products are kept in stock and marketed prominently.

```python
%%sql
WITH tot_sales_product AS (
    SELECT c.category_name, p.product_name, SUM(od.unit_price*od.quantity*(1-od.discount)) AS tot_sales
      FROM products AS p
      JOIN order_details AS od
        ON p.product_id=od.product_id
      JOIN categories AS c
        ON p.category_id=c.category_id
     GROUP BY c.category_name, p.product_name
),
prod_ranks AS (
    SELECT category_name, product_name, tot_sales, ROW_NUMBER() OVER(PARTITION BY category_name ORDER BY tot_sales DESC) AS prod_rank
    FROM tot_sales_product
)

SELECT *
  FROM prod_ranks
 WHERE prod_rank<=3;
```

## Top 25 Percentile of Products by Sales

```python
%%sql
WITH total_sales AS (
    SELECT p.product_name,c.category_name, SUM(od.unit_price*od.quantity*(1-od.discount)) AS tot_sales
      FROM products AS p
      JOIN order_details AS od
        ON p.product_id=od.product_id
      JOIN categories AS c
        ON p.category_id=c.category_id
     GROUP BY c.category_name,p.product_name
),

pct_sales AS (
    SELECT product_name, category_name, tot_sales, PERCENT_RANK() OVER(ORDER BY tot_sales) AS pct_rank
      FROM total_sales
      )

SELECT product_name, category_name, tot_sales
  FROM pct_sales
 WHERE pct_rank>=0.75
 ORDER BY tot_sales DESC
```

From the above table we can see that `Cte de Blaye` is the highest-selling product for Northwind Traders with its total sales being `75%` more than the second highest product `Thringer Rostbratwurst`. From a strategy standpoint, The Management at Northwind traders should enhance the visibility of the above products in their respective category sections by improving their placement on the shelves and aisles. The visibility of these products should be enhanced on the the online marketplace as well. 

Since, these products are star performers, improving their visibility will drive more sales, resulting in increased revenues for Northwind Traders. 

## Cumulative Sales of Product Categories by Month

Next we will look at the trend of growth in sales for various product categories.

```python
query="""
WITH cat_sales_month AS (
    SELECT DATE_TRUNC('month',o.order_date) AS order_month, c.category_name, 
    SUM(od.unit_price*od.quantity*(1-od.discount)) AS tot_sum
      FROM orders AS o
      JOIN order_details AS od
        ON o.order_id=od.order_id
      JOIN products AS p
        ON od.product_id=p.product_id
      JOIN categories AS c
        ON p.category_id=c.category_id
     GROUP BY order_month, c.category_name
     ORDER BY order_month
),

cat_sales_pivot AS (
    SELECT order_month, 
       SUM(CASE WHEN category_name='Beverages' THEN tot_sum ELSE 0 END) AS Beverages,
       SUM(CASE WHEN category_name='Condiments' THEN tot_sum ELSE 0 END) AS Condiments,
       SUM(CASE WHEN category_name='Confections' THEN tot_sum ELSE 0 END) AS Confections,
       SUM(CASE WHEN category_name='Dairy Products' THEN tot_sum ELSE 0 END) AS Dairy_Products,
       SUM(CASE WHEN category_name='Grains/Cereals' THEN tot_sum ELSE 0 END) AS Grains_or_Cereals,
       SUM(CASE WHEN category_name='Meat/Poultry' THEN tot_sum ELSE 0 END) AS Meat_or_Poultry,
       SUM(CASE WHEN category_name='Produce' THEN tot_sum ELSE 0 END) AS Produce,
       SUM(CASE WHEN category_name='Seafood' THEN tot_sum ELSE 0 END) AS Seafood
    FROM cat_sales_month
    GROUP BY order_month
)

SELECT order_month,SUM(Beverages) OVER(ORDER BY order_month) AS roll_beverages,
                   SUM(Condiments) OVER(ORDER BY order_month) AS roll_condiments,
                   SUM(Confections) OVER(ORDER BY order_month) AS roll_confections,
                   SUM(Dairy_Products) OVER(ORDER BY order_month) AS roll_dairy_products,
                   SUM(Grains_or_Cereals) OVER(ORDER BY order_month) AS roll_grains_or_cereals,
                   SUM(Meat_or_Poultry) OVER(ORDER BY order_month) AS roll_meat_or_poultry,
                   SUM(Produce) OVER(ORDER BY order_month) AS roll_produce,
                   SUM(Seafood) OVER(ORDER BY order_month) AS roll_seafood
  FROM cat_sales_pivot;
"""

results= %sql $query
cat_sales_month=pd.DataFrame(results)
cat_sales_month.set_index('order_month',inplace=True)
cat_sales_month

plt.plot(cat_sales_month['roll_beverages'], label="Beverages")
plt.plot(cat_sales_month['roll_condiments'], label="Condiments")
plt.plot(cat_sales_month['roll_confections'],label="Confections")
plt.plot(cat_sales_month['roll_dairy_products'], label="Dairy Products")
plt.plot(cat_sales_month['roll_grains_or_cereals'], label="Grains or Cereals")
plt.plot(cat_sales_month['roll_meat_or_poultry'], label="Meat or Poultry")
plt.plot(cat_sales_month['roll_produce'], label="Produce")
plt.plot(cat_sales_month['roll_seafood'],label="Seafood")
plt.gca().yaxis.set_major_formatter('${x:,.2f}')
plt.title("Cumulative Sales of Product Categories by Month")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.legend()
plt.grid()
plt.show()
                
```

From the above plot, it can be seen that `Beverages`, `Dairy Products`, `Confections` are the top 3 categories to have shown the highest cumuilative growth rate over the entire time period. From a strategy standpoint, The Management should use the above analysis to forecast the cumulative growth rates into the future and also increase the product offerings in the categories which show high rates of increase in Cumulative Sales.
